```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# mathe & science marks
x=np.array([[95,90],[85,82],[60,55],[75,75],[65,65],[81,56],[99,89],[47,55],[30,33],[40,50],[72,62],[69,89],[79,36],[62,78],[55,66]])
```

```python
km=KMeans(n_clusters=3,random_state=42)
km.fit(x)
```

```python
xyz=km.labels_ # each cluster belongs to which data
abc=km.cluster_centers_ # means of all data
```

```python
# Plot Graph
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],c=xyz,cmap='Set1')
plt.scatter(abc[:,0],abc[:,0],c='blue',marker='o')
plt.title('Student Report')
plt.xlabel('Maths Marks')
plt.ylabel('Science Marks')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
  # Temperature data -- max & min
x=np.array([[35,20],[40,23],[32,19],[50,30],[52,33],[10,3],[24,12],[8,0],[7,-2],[3,0]])       
```

```python
plt.scatter(x[:,0],x[:,1],color='blue')
plt.xlabel('Highest temp')
plt.ylabel('Lowest temp')
plt.title('City temp')
plt.show()
```

```python
abc=AgglomerativeClustering(n_clusters=3)
```

```python
prediction=abc.fit_predict(x)
```

```python
prediction
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=abc,palette='Oranges')
plt.xlabel('Highest temp')
plt.ylabel('Lowest temp')
plt.title('HC of temp')
plt.show()
```

```python
# Dendogram
li=linkage(x,method='ward')
plt.figure(figsize=(5,4))
dendrogram(li,orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.title('Temperature')
plt.show()
```
