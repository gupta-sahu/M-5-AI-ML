```python
from sklearn.metrics import precision_score , recall_score , f1_score
```

```python
real=[1,0,1,1,0,1]
predicted=[1,0,1,0,0,1]  #80% prediction from the models

precision=precision_score(real,predicted)  #All prediction , how many were actully positive correct
recall=recall_score(real,predicted)   #All actully positive correct , how many were correctly found
f1=f1_score(real,predicted)

print("Precision score :",precision)
print("Recall score is :",recall)
print("F1 score / Harmonic mean/ Average of Real & Predicted :" , f1)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score , recall_score , f1_score
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic=titanic[['survived', 'pclass', 'age',]].dropna()
```

```python
titanic
```

```python
x=titanic[['pclass', 'age']]
y=titanic['survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x=LogisticRegression()
```

```python
x
```

```python
x.fit(x_train,y_train)
```

```python
prediction=x.predict(x_test)
```

```python
print(prediction)
```

```python
# Evaluation

accuracy=accuracy_score(y_test,prediction)
precision=precision_score(y_test,prediction)
recall=recall_score(y_test,prediction)
f1=f1_score(y_test,prediction)

```

```python
print("Accuracy is :",accuracy)
print("Precision is :",precision)
print("Recall is ;",recall)
print("f1_score is :",f1)
```
