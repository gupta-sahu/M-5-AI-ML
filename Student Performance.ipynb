```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(4,3))
sns.histplot(df['math score'],bins=10,kde=True,color='pink')
plt.xlabel('Math Marks')
plt.ylabel('Frequency')
plt.title('Maths marks Dist.')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x=df['reading score'],y=df['writing score'],color='orange')
plt.xlabel('Reading Score')
plt.ylabel('Writing core')
plt.title('Maths Boxlplot.')
plt.show()
```

```python
df_box=df.head(20)
sns.boxplot(x=df_box['reading score'],y=df_box['writing score'],color='orange')
plt.xlabel('Reading Score')
plt.ylabel('Writing core')
plt.title('Maths Boxlplot.')
plt.show()
```

```python
xyz=df.select_dtypes(include=['number'])
sns.heatmap(xyz.corr(),annot=True,color='red')
plt.title('Correlation b/w variable')
plt.show()
```
